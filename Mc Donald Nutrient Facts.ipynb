```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Load the dataset
url = 'menu.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.columns
```

```python
data['Cluster'].value_counts()
```

```python
data = data.drop(['Category', 'Item', 'Serving Size'], axis=1)
data = data.fillna(0)
data_norm = (data - data.mean()) / data.std()
```

```python
data.head()
```

```python
# Perform K-means clustering
k = 5  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data_norm)
```

```python
# Analyze the clustering results
cluster_labels = kmeans.labels_
data['Cluster'] = cluster_labels
cluster_centers = kmeans.cluster_centers_
print('Cluster Centers:')
print(cluster_centers)
```

```python
# Predict the cluster for new data points
new_data = pd.DataFrame({
    'Calories': np.random.randint(0, 500, size=1),
    'Calories from Fat': np.random.randint(0, 500, size=1),
    'Total Fat': np.random.randint(0, 50, size=1),
    'Total Fat (% Daily Value)': np.random.randint(0, 100, size=1),
    'Saturated Fat': np.random.randint(0, 20, size=1),
    'Saturated Fat (% Daily Value)': np.random.randint(0, 100, size=1),
    'Trans Fat': np.random.randint(0, 10, size=1),
    'Cholesterol': np.random.randint(0, 300, size=1),
    'Cholesterol (% Daily Value)': np.random.randint(0, 100, size=1),
    'Sodium': np.random.randint(0, 3000, size=1),
    'Sodium (% Daily Value)': np.random.randint(0, 100, size=1),
    'Carbohydrates': np.random.randint(0, 100, size=1),
    'Carbohydrates (% Daily Value)': np.random.randint(0, 100, size=1),
    'Dietary Fiber': np.random.randint(0, 50, size=1),
    'Dietary Fiber (% Daily Value)': np.random.randint(0, 100, size=1),
    'Sugars': np.random.randint(0, 50, size=1),
    'Protein': np.random.randint(0, 50, size=1),
    'Vitamin A (% Daily Value)': np.random.randint(0, 100, size=1),
    'Vitamin C (% Daily Value)': np.random.randint(0, 100, size=1),
    'Calcium (% Daily Value)': np.random.randint(0, 100, size=1),
    'Iron (% Daily Value)': np.random.randint(0, 100, size=1)
})
new_data_norm = (new_data - data.mean()) / data.std()
```

```python
new_data_norm.columns
```

```python
new_data_norm.drop(['Cluster'], axis = 1, inplace = True)
```

```python
# Predict the cluster for new data points
new_data['Cluster'] = kmeans.predict(new_data_norm.values)
print('Predicted Cluster:')
print(new_data['Cluster'])
```
